## Stanモデルの記述

データブロック　→　パラメータブロック　→　統計モデルブロック

```python
stan_model = """
data {
  int N;
  real Y[N];
}

parameters {
  real mu;
  real<lower=0> sigma;
}

model {
  for (i in 1:N){
    Y[i] ~ normal(mu, sigma);
  }
}
"""
```

```python
sm = pystan.StanModel(model_code=stan_model)  # コンパイル
```

```python
stan_data = {"N":df.shape[0], "Y":df[0]}  # わたすデータを辞書型で定義
```

```python
fit = sm.sampling(data=stan_data, iter=2000, chains = 3, warmup= 500, seed=123)  # サンプリング（条件指定）　イテレーションは小さく初めて
```
